Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

#Answer

Matplotlib is a popular Python library for creating static, animated, and interactive visualizations. It provides a wide range of plotting capabilities and is extensively used for data exploration, analysis, and presentation.

Matplotlib is used for various purposes, including:

1. Data Visualization: Matplotlib allows you to create a wide variety of plots and charts to represent data in a visually appealing manner. It provides a high degree of customization, allowing you to control almost every aspect of the plot, such as colors, labels, titles, axes, and more.

2. Data Analysis: Matplotlib can be used for exploratory data analysis to gain insights from data. It helps in understanding the distribution, patterns, and relationships in the data by creating histograms, scatter plots, and other statistical plots.

3. Presentation and Reporting: Matplotlib enables the creation of publication-quality figures and graphics for scientific papers, presentations, and reports. It provides precise control over the visual elements, allowing you to create visually appealing and informative plots.

4. Integration with Other Libraries: Matplotlib integrates seamlessly with other scientific computing libraries in Python, such as NumPy and Pandas. It can plot data directly from these libraries' data structures and arrays, making it a versatile tool for data analysis.

5. Customization and Extensibility: Matplotlib offers extensive customization options, allowing you to tailor plots to your specific requirements. It provides a highly flexible and extensible framework, enabling the creation of specialized plots and visualizations.

Five plots that can be created using the Pyplot module of Matplotlib are:

1. Line Plot: A line plot shows the relationship between two variables by connecting data points with straight lines. It is commonly used to display trends over time or ordered categories.

2. Scatter Plot: A scatter plot represents individual data points as markers on a Cartesian plane. It is useful for visualizing the relationship between two continuous variables and identifying patterns or clusters in the data.

3. Bar Chart: A bar chart displays categorical data using rectangular bars with lengths proportional to the values they represent. It is commonly used to compare and present data across different categories.

4. Histogram: A histogram is a graphical representation of the distribution of numerical data. It divides the data into bins and displays the frequency or count of data points falling into each bin.

5. Pie Chart: A pie chart represents data as slices of a circular pie, with each slice representing a different category or proportion of a whole. It is useful for displaying relative proportions or percentages of a dataset.

These are just a few examples of the many types of plots that can be created using the Pyplot module of Matplotlib. The library provides a rich set of functions and options to create a wide range of visualizations.

Q2: What is a scatter plot? Use the following code to generate data for x and y.

Using this generated data plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

```python
#Answer

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

#Answer

The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to plot multiple subplots in a single figure window, organizing them in rows and columns.

Each subplot can be individually customized, and you can specify the number of rows, the number of columns, and the position of each subplot within the grid. The subplot() function takes three arguments: the number of rows, the number of columns, and the index of the subplot.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line plots
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating subplots
plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title('Line Plot 1')

plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title('Line Plot 2')

plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title('Line Plot 3')

plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title('Line Plot 4')

# Adjusting the layout
plt.tight_layout()

# Displaying the plots
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

#Answer

A bar plot, also known as a bar chart or bar graph, is a visualization that represents categorical data with rectangular bars. The length or height of each bar corresponds to the value or frequency of the category it represents. Bar plots are commonly used to compare and display the values of different categories, showing the relationships and variations between them.

Bar plots are useful for:

> Comparing Data: Bar plots allow easy visual comparison of values across different categories. They make it simple to identify the highest or lowest values and observe the relative differences between categories.

> Presenting Data Distribution: Bar plots can show the distribution of data in different categories, providing insights into the spread and variation within each category.

> Categorical Analysis: Bar plots help in analyzing categorical data by summarizing and visualizing the values associated with each category.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.bar(company, profit)
plt.title('Bar Plot')
plt.xlabel('Company')
plt.ylabel('Profit')

plt.show()

# Horizontal bar plot
plt.barh(company, profit)
plt.title('Horizontal Bar Plot')
plt.xlabel('Profit')
plt.ylabel('Company')

plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

#Answer

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays the summary statistics of a dataset, such as the minimum, first quartile, median (second quartile), third quartile, and maximum values. It provides insights into the central tendency, spread, and skewness of the data.

A box plot is used for several purposes:

> Data Distribution: A box plot visually represents the spread and distribution of the dataset. It provides a compact summary of the key statistical measures and displays any outliers or extreme values.

> Comparison: Box plots are useful for comparing the distributions of different datasets or groups. By placing multiple box plots side by side, you can easily compare their medians, quartiles, and ranges.

> Identifying Skewness and Outliers: Box plots help in identifying the skewness of the data distribution. Skewness indicates whether the data is skewed to the left or right. Outliers, which are values that deviate significantly from the overall pattern, are also visible in a box plot.

> Detecting Dispersion: Box plots provide information about the variability and dispersion of the data. The length of the box (interquartile range) represents the spread of the middle 50% of the data, while the whiskers show the range of the data within a certain threshold.

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.boxplot(data)
plt.title('Box Plot')
plt.xlabel('Box')
plt.ylabel('Values')

plt.show()
```
